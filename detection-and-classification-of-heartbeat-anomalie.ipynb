```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, f1_score
import pywt
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout
from keras.utils import to_categorical
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest, f_classif

# Veriyi yükle
data = pd.read_csv('/kaggle/input/heartbeat/mitbih_train.csv', header=None)

# Verinin ilk birkaç satırını görüntüle
print(data.head())

# Veriyi ve etiketleri ayır
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values

# Verinin boyutlarını kontrol et
print(f"Veri boyutu: {X.shape}")
print(f"Etiket boyutu: {y.shape}")

# Normalizasyon işlemi
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)

# Normalizasyon sonrası veri kontrolü
print(f"Normalleştirilmiş veri: {X_normalized[:5]}")

def extract_features(signal, wavelet='db1'):
    coeffs = pywt.wavedec(signal, wavelet)
    features = np.concatenate(coeffs)
    return features

# Tüm verisetinde özellik çıkarma
X_features = np.array([extract_features(signal) for signal in X_normalized])

print(f"Özellik çıkarılmış veri boyutu: {X_features.shape}")

# PCA ile boyut azaltma
pca = PCA(n_components=50)
X_pca = pca.fit_transform(X_features)

# Ek özellik çıkarma (ANOVA)
selector = SelectKBest(f_classif, k=50)
X_new = selector.fit_transform(X_pca, y)

# Veriyi eğitim ve test olarak ayırma
X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.2, random_state=42)

# Hiperparametre optimizasyonu
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10]
}

# GridSearchCV kullanarak en iyi hiperparametreleri bulma
grid_search = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=3, n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)

best_rf = grid_search.best_estimator_

# Modeli test etme
y_pred = best_rf.predict(X_test)

# Model performansını değerlendirme
accuracy = accuracy_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, best_rf.predict_proba(X_test), multi_class='ovr')
f1 = f1_score(y_test, y_pred, average='weighted')
report = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Doğruluk: {accuracy}")
print(f"ROC-AUC: {roc_auc}")
print(f"F1 Skoru: {f1}")
print(f"Sınıflandırma Raporu:\n{report}")
print(f"Karışıklık Matrisi:\n{conf_matrix}")

# Karışıklık matrisinin görselleştirilmesi
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues")
plt.xlabel('Tahmin Edilen')
plt.ylabel('Gerçek')
plt.title('Karışıklık Matrisi')
plt.show()

# LSTM modelinin oluşturulması
y_train_encoded = to_categorical(y_train)
y_test_encoded = to_categorical(y_test)

model_dl = Sequential()
model_dl.add(LSTM(64, input_shape=(X_train.shape[1], 1), return_sequences=True))
model_dl.add(Dropout(0.2))
model_dl.add(LSTM(64, return_sequences=False))
model_dl.add(Dropout(0.2))
model_dl.add(Dense(y_train_encoded.shape[1], activation='softmax'))

model_dl.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Modelin eğitilmesi
X_train_reshaped = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test_reshaped = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

model_dl.fit(X_train_reshaped, y_train_encoded, epochs=10, batch_size=64, validation_split=0.2)

# Modelin test edilmesi
y_pred_dl = model_dl.predict(X_test_reshaped)
y_pred_dl_classes = np.argmax(y_pred_dl, axis=1)

# Performans değerlendirme
accuracy_dl = accuracy_score(y_test, y_pred_dl_classes)
roc_auc_dl = roc_auc_score(y_test, y_pred_dl, multi_class='ovr')
f1_dl = f1_score(y_test, y_pred_dl_classes, average='weighted')
report_dl = classification_report(y_test, y_pred_dl_classes)
conf_matrix_dl = confusion_matrix(y_test, y_pred_dl_classes)

print(f"Doğruluk (DL): {accuracy_dl}")
print(f"ROC-AUC (DL): {roc_auc_dl}")
print(f"F1 Skoru (DL): {f1_dl}")
print(f"Sınıflandırma Raporu (DL):\n{report_dl}")
print(f"Karışıklık Matrisi (DL):\n{conf_matrix_dl}")

plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix_dl, annot=True, fmt="d", cmap="Blues")
plt.xlabel('Tahmin Edilen')
plt.ylabel('Gerçek')
plt.title('Karışıklık Matrisi (DL)')
plt.show()
```
